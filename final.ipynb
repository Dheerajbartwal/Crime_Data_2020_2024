```python
import seaborn as sns

# Load the dataset
data = pd.read_csv('cleaned_file.csv')

# Set up plot style
plt.style.use('seaborn-darkgrid')

# 1. Crime Frequency by Area (Bar Plot)
plt.figure(figsize=(12, 6))
area_counts = data['AREA NAME'].value_counts()
sns.barplot(x=area_counts.index, y=area_counts.values, palette='viridis')
plt.title('Crime Frequency by Area', fontsize=16)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Area Name', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Top 15 Most Common Crime Types (Bar Plot)

plt.figure(figsize=(12, 6))
crime_counts = data['Crm Cd Desc'].value_counts()[:15]
sns.barplot(y=crime_counts.index, x=crime_counts.values, palette='coolwarm')
plt.title('Top 15 Most Common Crime Types', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Crime Type', fontsize=12)
plt.show()
```

```python
# 3. Victim Age Distribution (Histogram)

plt.figure(figsize=(10, 6))
sns.histplot(data['Vict Age'], bins=30, kde=True, color='orange')
plt.title('Victim Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# 4. Crime Occurrences by Time of Day (Line Plot)
plt.figure(figsize=(12, 6))
time_counts = data['TIME OCC'].value_counts().sort_index()
sns.lineplot(x=time_counts.index, y=time_counts.values, color='green')
plt.title('Crime Occurrences by Time of Day', fontsize=16)
plt.xlabel('Time (24-Hour Clock)', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.show()
```

```python
# 5. Victim Gender Distribution (Pie Chart)
plt.figure(figsize=(8, 8))
gender_counts = data['Vict Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'pink', 'lightgray'])
plt.title('Victim Gender Distribution', fontsize=16)
plt.show()
```

```python

# 6. Case Status Distribution (Pie Chart)
plt.figure(figsize=(8, 8))
status_counts = data['Status Desc'].value_counts()
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2', len(status_counts)))
plt.title('Case Status Distribution', fontsize=16)
plt.show()
```

```python
# 7. Crime Frequency by Premises Description (Bar Plot)
plt.figure(figsize=(12, 6))
premis_counts = data['Premis Desc'].value_counts()[:15]
sns.barplot(y=premis_counts.index, x=premis_counts.values, palette='magma')
plt.title('Top 15 Premise Descriptions for Crimes', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Premise Description', fontsize=12)
plt.show()
```
